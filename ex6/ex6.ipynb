```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

import librosa
import librosa.display
import sounddevice as sd
```

```python
audio, sr = librosa.load('brahms_hungarian_dance_5_short.wav')
#audio, sr = librosa.load('classic_rock_beat.wav')
#audio, sr = librosa.load('latin_groove_short.wav')
#audio, sr = librosa.load('conga_groove.wav')

hop_length = 1024

# Plotting spectogram 
dft = librosa.stft(audio)  # STFT of y
S_db = librosa.amplitude_to_db(np.abs(dft), ref=np.max)
plt.figure()
librosa.display.specshow(S_db)
plt.colorbar()

# Calculating constant-Q spectrogram 
chroma_cqt = librosa.feature.chroma_cqt(y=audio,
                                        hop_length=hop_length, 
                                        sr=sr, tuning=0, 
                                        norm=2)
plt.figure()
librosa.display.specshow(chroma_cqt, y_axis='chroma', x_axis='time',)
plt.colorbar()

# Calculating chromagram 
chromagram = librosa.feature.chroma_stft(audio, sr)
plt.figure()
librosa.display.specshow(chromagram, y_axis='chroma',x_axis='time',)
plt.colorbar()
```
